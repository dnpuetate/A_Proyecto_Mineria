```python
#Importar librerías 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import svm
import seaborn as sn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import export_text
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
import seaborn as sns
sns.set_style('darkgrid')

# importar clases extra para modelar
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

# multivariate imputation
from sklearn.impute import KNNImputer

# the scaler - for standardisation
from sklearn.preprocessing import StandardScaler

#Mostrar todas las columnas del dataset usando la funcion max_columns
pd.set_option('display.max_columns', None)
```

# Load data

```python
# Importar dataset
def subir():
    df = pd.read_csv("train.csv", index_col="date", parse_dates=True)
    return df
```

```python
#Mostrar los primeros n-registros del dataset 
df=subir()
print('Forma:', df.shape)
df.head()
```

```python
#Familias de productos
df = subir()
family_unique = len(df.family.unique())
family_unique
```

```python
#Etiquetar los datos 

df=subir()
ohe = OneHotEncoder(sparse=False, handle_unknown='ignore')
dfOh = pd.DataFrame( ohe.fit_transform( np.array(df.family).reshape( (-1, 1) ) ) )
dfOh.columns = df.family.unique()
dfOh.index = df.index

df = pd.concat([df.drop("family", axis=1), dfOh], axis=1)
df.head()

```

```python
plt.plot(df.groupby(df.index)["sales"].mean())
```

```python
# Separar el conjunto de prueba y de text
feature_cols  = ['id', 'store_nbr','onpromotion', 'AUTOMOTIVE', 'BABY CARE', 
                 'BEAUTY', 'BEVERAGES', 'BOOKS', 'BREAD/BAKERY',
                 'CELEBRATION', 'CLEANING', 'DAIRY', 'DELI', 'EGGS', 'FROZEN FOODS', 
                'GROCERY I', 'GROCERY II', 'HARDWARE', 'HOME AND KITCHEN I', 
                 'HOME AND KITCHEN II', 'HOME APPLIANCES', 'HOME CARE',
                 'LADIESWEAR', 'LAWN AND GARDEN', 'LINGERIE',
                 'LIQUOR,WINE,BEER', 'MAGAZINES', 'MEATS', 'PERSONAL CARE',
                 'PET SUPPLIES', 'PLAYERS AND ELECTRONICS', 'POULTRY', 'PREPARED FOODS', 
                 'PRODUCE', 'SCHOOL AND OFFICE SUPPLIES', 'SEAFOOD', 'sales'
                ]

# Soltar el objetivo de la lista de características
feature_cols.remove('sales')


X_train, X_test, y_train, y_test = train_test_split(
    df[feature_cols],  # solo las caracteristicas (features)
    df['sales'],  # el objetivo (target)
    test_size=0.3,  # el porcentaje de obs en el conjunto de prueba
    random_state=0)  # for reproducibility

X_train.shape, X_test.shape
```

```python
# restablecer el índice, para que podamos comparar los valores más adelante
# en la demostración

X_train.reset_index(inplace=True, drop=True)
X_test.reset_index(inplace=True, drop=True)
```

```python
#LIMPIAR LOS DATOS
# 1. Encontrar variables con datos faltantes
def falta(df):
    for var in df.columns:
        if df[var].isnull().sum() > 1:
            print(var, df[var].isnull().sum())
```

# KNN imputation

```python
imputer = KNNImputer(
    n_neighbors=5, # the number of neighbours K
    weights='distance', # the weighting factor
    metric='nan_euclidean', # the metric to find the neighbours
    add_indicator=False, # whether to add a missing indicator
)
```

```python

imputer.fit(X_train)
```

```python
train_t = imputer.transform(X_train)
test_t = imputer.transform(X_test)

# sklearn returns a Numpy array
# lets make a dataframe
train_t = pd.DataFrame(train_t, columns=X_train.columns)
test_t = pd.DataFrame(test_t, columns=X_test.columns)

train_t.head()
```

# Standardisation 

```python
#Describir los datos
df.describe()
```

```python
# estandarización: con StandardScaler de sklearn

# configurar la escala
scaler = StandardScaler()
# Ajuste el escalador al train, aprenderá los parámetros
scaler.fit(X_train)

# transformar en conjuntos de train y test
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# el escalador almacena la media de las características, 
#aprendidas del conjunto de entrenamiento
scaler.mean_
```

```python
# el escalador almacena la desviación estándar de las características,
# aprendido del conjunto de entrenamiento

scaler.scale_
```

```python
# transformemos las matrices NumPy devueltas en marcos de datos para el resto de
# la demostración

X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
# echemos un vistazo al conjunto de datos de entrenamiento original: media y desviación estándar
# Uso np.round para reducir el número de decimales a 1

np.round(X_train.describe(), 1)
```

```python
# echemos un vistazo al conjunto de datos de entrenamiento escalado: media y desviación estándar
# Uso np.round para reducir el número de decimales a 1.

np.round(X_train_scaled.describe(), 1)
```
